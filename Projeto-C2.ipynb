```python
import cmath
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Eq2:
    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c
        self.delta = b**2 - 4*a*c
    
    def raizes(self):
        if self.delta == 0:
            return [-self.b /(2*self.a),-self.b /(2*self.a)]
        elif self.delta < 0:
            sol1 = (-self.b + cmath.sqrt(self.delta))/(2*self.a)
            sol2 = (-self.b - cmath.sqrt(self.delta))/(2*self.a)
            return [sol1, sol2]
        else:
            sol1 = (-self.b + math.sqrt(self.delta))/(2*self.a)
            sol2 = (-self.b - math.sqrt(self.delta))/(2*self.a)
            return [sol1, sol2]
    
    def grafico(self):
        r = self.raizes()
        x =np.linspace(r[0]-2, r[1]+2, 40)
        y= self.a * x**2 + self.b*x + self.c
        
        plt.plot(x, y)
        plt.show()
```

```python
# %%
eq = Eq2(1,0,0)
```

```python
eq.grafico()
```
